```python
from qiskit import QuantumCircuit
```

```python
import math
import numpy as np
from qiskit.quantum_info import Operator

qc_ms = QuantumCircuit(3)
qc_ms.ms(math.pi / 2, [0,2])
print(qc_ms)
mat_ms = Operator(qc_ms).to_matrix()

qc_rxx = QuantumCircuit(3)
qc_rxx.rxx(math.pi / 2, 0,2)
print(qc_rxx)
mat_rxx = Operator(qc_rxx).to_matrix()

print(np.allclose(mat_ms, mat_rxx))
```
